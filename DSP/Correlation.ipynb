```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cross-Correlation

# x = [2,3,1,5]
# h = [1,2,3]
x_t = input("Enter the Input Sequence: ")
x = [int(n) for n in x_t.split()]
h_t = input("Enter the Impulse Response Sequence: ")
h = [int(n) for n in h_t.split()]
L = len(x)+len(h)-1
y = np.array([0]*L)
hh = np.flip(h)
for i in range(len(x)) :
    for j in range(len(h)) :
        y[i+j] = y[i+j]+x[i]*hh[j]
        

plt.stem(y)
y
```

```python
np.correlate(x,h,'full')
```
