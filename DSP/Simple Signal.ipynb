```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constant Signal
L = 10
s = [0]*L
for n in range(L) :
    s[n] = 5
plt.stem(s)
```

```python
# Constant Signal
L = 10
s = [5]*L
x = np.arange(-5,5)
plt.stem(x,s)
```

```python
# Unit Step Signal
L = 10
s = [1]*L
for n in range(5) :
    s[n] = 0
x = np.arange(-5,5)
plt.stem(x,s)
```

```python
# Unit Sequence Signal
L = 20
s = [0]*L
s[10] = 1
x = np.arange(-10,10)
plt.stem(x,s)
```

```python
# Unit Ramp Signal
L = 20
s = [0]*L
for n in range(L):
    s[n] = n
plt.stem(s)
```

```python
# Exponential Signal

L = 200
a = 0.9
s = [0]*L
for n in range(L) :
    s[n] = a**n
plt.stem(s)
    
```
