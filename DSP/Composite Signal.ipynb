```python
# Composite Signal of 3 simple signals
import numpy as np
import matplotlib.pyplot as plt
f1 = 4
f2 = 10
f3 = 14
Fs = 50
L = 100
s1 = [0]*L
s2 = [0]*L
s3 = [0]*L
s = [0]*L
for n in range(L) :
    s1[n] = np.sin(2*np.pi*f1*n/Fs)
for n in range(L) :
    s2[n] = np.sin(2*np.pi*f2*n/Fs)
for n in range(L) :
    s3[n] = np.sin(2*np.pi*f3*n/Fs)
for n in range(L) :
    s[n] = s1[n]+s2[n]+s3[n]
plt.plot(s)
```
