```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = [0,0,0,2,3,4,5,0,0,0] 
n = np.arange(-5,5) 
plt.stem(n,x)
n,x
```

```python
# Right Shift to 1

L = len(x)
y = [0]*(L+1)
for i in range(L) :
    y[i+1] = x[i]
plt.stem(y)
x,y
```

```python
# Left Shift to 2

L = len(x)
y = [0]*(L+1)
for i in range(L) :
    y[i-2] = x[i]
plt.stem(y)
x,y
```

```python
# Right Shift to 2
L = len(x)
y = [0]*L
for i in range(L-1,0,-1) :
    y[i] = x[i-2]
plt.stem(y)
x,y
```

```python
# Left Shift to 1
L = len(x)
y = [0]*L
for i in range(L-2) :
    y[i] = x[i+1]
plt.stem(y)
x,y
```

```python
# Folding
L = len(x)
y = [0]*L
n = np.arange(-5,5)
for i in range(L) :
    y[i] = x[-i]
plt.stem(n,y)
n,y
```
